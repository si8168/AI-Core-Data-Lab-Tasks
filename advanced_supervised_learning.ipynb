## Binary Classification — Breast Cancer Diagnosis Prediction

### 1. Problem Identification & Machine Learning Type
* **The Clinical Problem:** Predicting whether a tumor is malignant (cancerous) or benign (non-cancerous) based on structural cell measurements.
* **Machine Learning Type:** This is a **Supervised Binary Classification** task. The output consists of exactly two discrete, mutually exclusive classes (`0` for Malignant, `1` for Benign).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer, fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix, 
                             mean_absolute_error, mean_squared_error, r2_score)

print("==================================================================")
print("PART 1: BINARY CLASSIFICATION - BREAST CANCER DIAGNOSIS PREDICTION")
print("==================================================================\n")

# 1. Load dataset
cancer = load_breast_cancer()
X_cancer = pd.DataFrame(cancer.data, columns=cancer.feature_names)
y_cancer = cancer.target

print(f"Dataset Loaded Successfully.")
print(f"Features Shape: {X_cancer.shape}")
print(f"Target Classes: {cancer.target_names} (0 = Malignant, 1 = Benign)\n")

# 2. Train-Test Split (80/20 Stratified)
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X_cancer, y_cancer, test_size=0.2, random_state=42, stratify=y_cancer
)

# 3. Scaling Features
scaler_c = StandardScaler()
X_train_c_scaled = scaler_c.fit_transform(X_train_c)
X_test_c_scaled = scaler_c.transform(X_test_c)

# 4. Train & Evaluate Models
clf_models = {
    "Support Vector Classifier (SVC)": SVC(kernel='linear', random_state=42),
    "Random Forest Classifier": RandomForestClassifier(random_state=42)
}

for name, model in clf_models.items():
    model.fit(X_train_c_scaled, y_train_c)
    predictions = model.predict(X_test_c_scaled)
    
    print(f"--- {name} Performance ---")
    print(f"Accuracy Score: {accuracy_score(y_test_c, predictions):.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test_c, predictions, target_names=cancer.target_names))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test_c, predictions), "\n")
```

### 2. Data Preprocessing & Leakage Mitigation
* **Class Balancing:** We implement an 80/20 train-test split using `stratify=y`. This forces the splitting algorithm to maintain identical proportions of malignant and benign samples across both subsets.
* **Standard Scaling:** To avoid catastrophic data leakage, features are transformed using a `StandardScaler`. The scaling parameters are mathematically computed from the training chunk only and subsequently mapped onto the testing data.

## Continuous Regression — California House Price Prediction

### 1. Problem Identification & Machine Learning Type
* **The Economic Problem:** Estimating regional median home values based on localized demographics, age of housing, and geographic coordinates.
* **Machine Learning Type:** This is a **Supervised Continuous Regression** task. Unlike classification, the target variable is a numerical trajectory along an infinite continuous spectrum (prices in hundreds of thousands of dollars).

```python
print("================================================================")
print("PART 2: CONTINUOUS REGRESSION - CALIFORNIA HOUSE PRICE PREDICTION")
print("================================================================\n")

# 1. Load dataset
housing = fetch_california_housing()
X_house = pd.DataFrame(housing.data, columns=housing.feature_names)
y_house = housing.target  # Expressed in hundreds of thousands of dollars ($100,000s)

print(f"Dataset Loaded Successfully.")
print(f"Features Shape: {X_house.shape}")
print(f"Target Variable: Median House Value (Continuous Variable)\n")

# 2. Train-Test Split (80/20)
X_train_h, X_test_h, y_train_h, y_test_h = train_test_split(
    X_house, y_house, test_size=0.2, random_state=42
)

# 3. Scaling Features
scaler_h = StandardScaler()
X_train_h_scaled = scaler_h.fit_transform(X_train_h)
X_test_h_scaled = scaler_h.transform(X_test_h)

# 4. Train & Evaluate Models
reg_models = {
    "Multiple Linear Regression": LinearRegression(),
    "Random Forest Regressor": RandomForestRegressor(n_estimators=50, random_state=42, n_jobs=-1)
}

for name, model in reg_models.items():
    model.fit(X_train_h_scaled, y_train_h)
    predictions = model.predict(X_test_h_scaled)
    
    # Metrics calculation
    mae = mean_absolute_error(y_test_h, predictions)
    mse = mean_squared_error(y_test_h, predictions)
    r2 = r2_score(y_test_h, predictions)
    
    print(f"--- {name} Performance ---")
    print(f"Mean Absolute Error (MAE):     {mae:.4f}")
    print(f"Mean Squared Error (MSE):      {mse:.4f}")
    print(f"R² Variance Score:             {r2:.4f}")
    print(f"Explanation: This model explains {r2*100:.2f}% of the house price variance.\n")
```

### 3. Metric Justification & Clinical Reflection
For medical binary diagnostics, **Accuracy alone is a dangerous metric**. We must heavily examine the **Confusion Matrix** and look at **Recall**. A false negative (classifying a malignant tumor as benign) means missing a patient's cancer, which is a critical error. We aim for a model that minimizes these specific clinical misses.

### Theoretical Comparison: Classification vs. Regression

This notebook demonstrates the dual pillars of Supervised Machine Learning: **Classification** and **Regression**.

1. **Classification (Breast Cancer Dataset):** Our goal was to assign discrete, categorical labels to inputs—specifically a binary choice between `malignant` (0) and `benign` (1). Performance was audited using a **Confusion Matrix** to track errors. In medical diagnosis, maximizing **Recall** (reducing false negatives) is vital, as misclassifying a malignant tumor as benign has severe consequences. 
   
2. **Regression (California Housing Dataset):** Our goal was to predict a continuous numerical value (median home values). Because the target outputs are infinite numerical points rather than categories, classification metrics are invalid. Instead, we applied **Mean Absolute Error (MAE)** to assess the average price deviation in dollars and the **$R²$ Coefficient of Determination** to calculate the percentage of real-world variance our models successfully mapped.
